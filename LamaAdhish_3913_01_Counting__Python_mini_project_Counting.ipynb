**How many outcomes were possible if you were not allowed to select an item multiple times.. Give two answers, with order mattering and not.**

Permutation is an arrangement in which the order is considered. To arrange in a particular order for 'r' objects from 'n' numbers of objects,

$$
nPr = \frac{n!}{(n-r)!}
$$

Therefore, if the order is taken into consideration, 10 students can be selected among from 20 students in the following ways:

```python
import math
n = 20
r = 10
print (math.perm(n,r))
```

How many outcomes were possible in your simulation if order did not matter?

Combination is the arrangement in which the order did not matter. 'r' objects can be choosen from 'n' objects in the following ways:

$$
nCr = \frac{n!}{n!(n-r)!}
$$

Therefore,if the order didn't matter, 10 students can be selected from among 20 students in the following ways:

```python
import math
n = 20
r = 10
print (math.comb(n,r))
```
